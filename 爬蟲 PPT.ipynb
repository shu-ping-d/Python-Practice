```python
import requests
```

```python
#網址
URL="https://www.ptt.cc/bbs/Gossiping/index.html"
#設定
my_headers={"cookie":"over18=1;"}
#發送請求到PPT八卦版
response=requests.get(URL,headers=my_headers)
print(response.text)
```

```python
#導入beautifulSoup 模組
import bs4
soup=bs4.BeautifulSoup(response.text,"html.parser")
```

```python
titles=soup.find_all("div","title")
```

```python
for t in titles:
    print(t.text)
```
